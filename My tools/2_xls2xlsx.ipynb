```python
import pandas as pd
import os
from tqdm.notebook import tqdm
import time
import re
```

```python
files = os.listdir('xls')
regex = re.compile(r"xls$")
files = list(filter(regex.search, files))
```

```python
bool_mask = [len(x) == 12 for x in files]
```

```python
files2 = [files[i] for i in range(len(files)) if bool_mask[i]]
```

```python
files = files2
```

```python
start = time.time()
errors = []
for file in tqdm(files):
    try:
        df = pd.read_excel('xls/' + file)
        df.to_excel('xlsx/' + file[0:8] + '.xlsx', index = False)
    except ValueError:
        errors.append(file)
print(str((time.time() - start)/60) + ' mins')
```
